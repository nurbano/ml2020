## Ejercicio 2
Se crean 2 data frame con distribución normal, con diferente ancho y centro.

```python
np.random.seed(1)

df1= pd.DataFrame(np.random.normal(scale=10, size=(50,2))+10,columns=["val", "etiq"])
df2= pd.DataFrame(np.random.normal(scale=5, size=(50,2))+20,columns=["val", "etiq"])
df1["etiq"]= False
df2["etiq"]= True
df_res = pd.concat([df1,df2], ignore_index=True)
```

```python
sns.pairplot(data=df_res, hue="etiq")
```

Se elige un umbral de acuerdo a una rápida observación gráfica de las distribuciones.

```python
umbral= 12
y_pred=df_res.val.apply(lambda val: val>12)
```

Se utiliza la librería sklearn.metrics para obtener la matriz de confusión. Y se calculan los parámetros que indican la calidad de la predicción.

```python
from sklearn.metrics import confusion_matrix
tn, fp, fn, tp= confusion_matrix(df_res.etiq, y_pred).ravel()
print("True Negative:\t",tn, "\nFalse Positive:\t",fp, "\nFalse Negative:\t",fn, "\nTrue Positive:\t",tp)
accuracy= (tp+tn)/(tp+tn+fp+fn)
print("Accuracy:\t", accuracy)
recall= tp/(tp+fn)
print("Recall:\t\t", recall)
precision= tp/(tp+fp)
print("Precision:\t", precision)
f_mesure= 2*precision*recall/(precision+recall)
print("F-Mesure:\t", f_mesure)
```

## Curva ROC
Se realizará la curva ROC para obtener el umbral óptimo.

Metodo:

1º Se ordenan los valores de menor a mayor.

2º Se comienza a barrer el array desde el final hacia el inicio.

3º Se crear un genera un array de 50x50, donde si el valor es positivo se coloca una recta vertical de 1/50 de altura, y si es falso se coloca una recta horizonal de 1/50 de ancho.

4º Se grafica el array, en donde se ovserva la curva roc.

5º Se grafica una recta y= 1-x.

6º Se encuentra el punto de intersección entre la curva roc y la recta, siendo este el punto óptimo.


```python
df_ord= df_res.sort_values("val", ignore_index=True)
df_ord
step=0.02
x=0
y=0
X=[0]
Y=[0]
for i in range (100):
    index= 99-i
    if df_ord.etiq[index]==True:
        X.append(x)
        Y.append(y+step)
        y+= step
        
    else:
        X.append(x+step)
        Y.append(y)
        x+= step  
x_r= np.array(X)  
y_r= 1-x_r
inter= np.array(Y)+y_r
plt.plot(X,Y)
plt.plot(x_r,y_r)
plt.grid()
idx = np.argwhere(np.diff(np.sign(Y - y_r))).flatten()+1
index= idx[0]
plt.plot(X[index], Y[index], 'ro')
plt.xlabel("FPR")
plt.ylabel("TPR")
print("FPR: ",X[index],"TPR: ",Y[index])
```

```python
a=df_ord.values[[index-1],0]
b=df_ord.values[[index],0]
a[0], b[0]
```

Se determina el umbral de acuerdo a la curva ROC

```python
sns.pairplot(data=df_res, hue="etiq")
plt.axvline(18.66)
umbral= (a[0]+b[0])/2
y_pred=df_res.val.apply(lambda val: val>umbral)
```

```python
tn, fp, fn, tp= confusion_matrix(df_res.etiq, y_pred).ravel()
print("True Negative:\t",tn, "\nFalse Positive:\t",fp, "\nFalse Negative:\t",fn, "\nTrue Positive:\t",tp)
accuracy= (tp+tn)/(tp+tn+fp+fn)
print("Accuracy:\t", accuracy)
recall= tp/(tp+fn)
print("Recall:\t\t", recall)
precision= tp/(tp+fp)
print("Precision:\t", precision)
f_mesure= 2*precision*recall/(precision+recall)
print("F-Mesure:\t", f_mesure)
```

Se modifica los dataframe, juntando los centros de las distribuciones

Se utiliza la curva ROC para determinar el valor óptimo

```python
df_ord2= df_res2.sort_values("val", ignore_index=True)
step=0.02
x=0
y=0
X=[0]
Y=[0]
for i in range (100):
    index= 99-i
    if df_ord2.etiq[index]==True:
        X.append(x)
        Y.append(y+step)
        y+= step
        
    else:
        X.append(x+step)
        Y.append(y)
        x+= step  
x_r= np.array(X)  
y_r= 1-x_r
inter= np.array(Y)+y_r
plt.plot(X,Y)
plt.plot(x_r,y_r)
plt.grid()
idx = np.argwhere(np.diff(np.sign(Y - y_r))).flatten()+1
index= idx[0]
plt.plot(X[index], Y[index], 'ro')
plt.xlabel("FPR")
plt.ylabel("TPR")
print("FPR: ",X[index],"TPR: ",Y[index])
```

```python
a=df_ord2.values[[index-1],0]
b=df_ord2.values[[index],0]
sns.pairplot(data=df_res2, hue="etiq")
plt.axvline((a[0]+b[0])/2)
umbral= 18.66
y_pred2=df_res2.val.apply(lambda val: val>umbral)
tn, fp, fn, tp= confusion_matrix(df_res2.etiq, y_pred2).ravel()
print("True Negative:\t",tn, "\nFalse Positive:\t",fp, "\nFalse Negative:\t",fn, "\nTrue Positive:\t",tp)
accuracy= (tp+tn)/(tp+tn+fp+fn)
print("Accuracy:\t", accuracy)
recall= tp/(tp+fn)
print("Recall:\t\t", recall)
precision= tp/(tp+fp)
print("Precision:\t", precision)
f_mesure= 2*precision*recall/(precision+recall)
print("F-Mesure:\t", f_mesure)
```
